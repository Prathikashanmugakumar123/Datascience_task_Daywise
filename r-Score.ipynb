```python
import math
Z = 1.96      # 95% confidence
sigma = 10    # population std dev
n = 100       # sample size
ME = Z * (sigma / math.sqrt(n))
print("Margin of Error =", ME)
```

```python
import numpy as np
# Example dataset
hours = np.array([1, 2, 3, 4, 5])
marks = np.array([30, 40, 50, 65, 80])
# Correlation matrix
r_matrix = np.corrcoef(hours, marks)
# Extract r value
r = r_matrix[0, 1]
print("Correlation coefficient (r):", r)
```
