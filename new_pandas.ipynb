```python
import pandas as pd

data = {
    "Name": ["Ram", "Sam", "Ravi"],
    "Age": [20, 21, 19],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)
df.to_csv("data.csv")
print(df)
```

```python
data = {
    "Department": ["IT", "IT", "CSE", "CSE"],
    "Salary": [50000, 60000, 55000, 65000]
}

df = pd.DataFrame(data)

print(df.groupby("Department").mean())
```
