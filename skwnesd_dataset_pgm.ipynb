```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("50_Startups.csv")
print(df)
```

```python
skew_value = df["Profit"].skew()
print("\nSkewness Value :", round(skew_value, 4))
```

```python
if skew_value > 0:
    print("Positive Skewed Distribution")

elif skew_value < 0:
    print("Negative Skewed Distribution")

else:
    print("Symmetrical Distribution")
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Profit"], bins=6)

plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.title("Monthly Profit Distribution")

plt.grid(True)

plt.show()
```
