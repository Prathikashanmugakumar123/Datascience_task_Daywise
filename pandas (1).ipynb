```python
import matplotlib.pyplot as plt
import numpy as np
 # Temperature over a week
days = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
temperature = np.array([30, 32, 31, 29, 35, 36, 34])
plt.plot(days, temperature, color='green', marker='o')
plt.title('Daily Temperature Over a Week')
plt.xlabel('Days')
plt.ylabel('Temperature (°C)')
plt.show()
```
