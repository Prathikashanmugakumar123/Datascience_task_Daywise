```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.read_csv("Iris.csv")

stats.probplot(
    df["SepalLengthCm"],
    dist="norm",
    plot=plt
)

plt.title("normal  distribution QQ Plot - Iris Sepal Length")
plt.show()
```
