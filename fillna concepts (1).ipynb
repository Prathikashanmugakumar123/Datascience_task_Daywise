```python
import pandas as pd

data = {
    "Age": [22, None, 29, None, 31],
    "City": ["Delhi", None, "Mumbai", None, "Chennai"],
    "Salary": [50000, 60000, None, None, 75000]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.fillna(0))
```

```python
print(df["Age"].fillna(df["Age"].mean()))
```

```python
print(df["Age"].fillna(df["Age"].median()))
```

```python
#Mode = Chennai / Delhi / Mumbai → All equal → Mode selects alphabetically → "Chennai"
print(df["City"].fillna(df["City"].mode()[0]))
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
#for checking isnull() occurs in dataset are not
print(df.fillna(0, inplace=True))
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Marks": [50, np.nan, np.nan, 70, np.nan, np.nan, 90],
    "City": ["Delhi", "Mumbai", np.nan, np.nan, "Chennai", np.nan, np.nan]
})

print("Original DataFrame:")
print(df)

df.fillna(method="ffill", limit=1)
```
